## Шаг 2. Предобработка данных

```python
#вычисляем проущенные значения
data.isna().sum()
```

```python
#вычисляем дубликаты
data.duplicated().sum()
```

```python
#переименовываем столбцы
data = data.rename(columns={'cityCenters_nearest':'city_centers_nearest','parks_around3000':'parks_around_3000','ponds_around3000':'ponds_around_3000'})
```

```python
#не во всех квартирах есть балконы, решено заменить пропущенные значения на 0
data['balcony'] = data['balcony'].fillna(0) 
data['balcony'].isna().sum()
```

```python
#не во всех домах есть апартаменты, решено их заменить на False, так как это object
data['is_apartment'] = data['is_apartment'].fillna(False) 
data['is_apartment'].isna().sum() 
```

```python
#заменяем пропуски в floors_total на значение floor (этаж в объявлении) + 1
data['floors_total'] = data['floors_total'].fillna(data['floor'] + 1)
data['floors_total'].isna().sum()
```

```python

#вычисляю зависимость цены от высоты потолков
print(data['ceiling_height'].describe())
data['a_price'] = data['last_price'] / data['total_area']
data[['ceiling_height', 'last_price', 'a_price']].corr() 
```

```python
#fig, axes = plt.subplots(1,2,figsize=(15,4))

#spb_center['ceiling_height'].hist(bins=30, ax=axes[0])
#data['ceiling_height'].hist(bins=30, ax=axes[1])

#axes[0].set_title('Распределение высоты потолков в центре')
#axes[1].set_title('Распределение высоты потолков по всему городу')

#plt.show()
```

Зависимость   высоты потолков от общей цены и цены квадратного метра не наблюдается, принимаю решение заменить все NaN и значения менее 2.7 в столбце ceiling_height на 2.7, а также все значения более 3.5 на 3.5, так как средняя высота потолков от застройщика в среднем такая.


```python

data['ceiling_height'].fillna(2.7, inplace=True) # заменяю пропуски в высоте потолков значением 2.7
data.loc[data['ceiling_height']<2.7,'ceiling_height']=2.7 # всё что ниже 2.7 заменяю на 2.7
data.loc[data['ceiling_height']>3.5,'ceiling_height']=3.5 # всё что выше 3.5 заменяю на 3.5
data['ceiling_height'].hist(bins=30)
data['ceiling_height'].isna().sum()
```

<br> Пропущенных значений в living_area = 1903.
<br> Наблюдается высокая степень корреляции living_area c total_area, rooms, last_price.
<br> В таблице наблюдается, что площадь кухни, больше жилой площади, решено менять значения местами.

```python
print('Количество Nan в living_area:', data['living_area'].isna().sum())
data['living_area'].describe()
data[['total_area', 'living_area', 'kitchen_area', 'rooms', 'last_price']].corr()
print('living_area < 18:', len(data[data['living_area'] < 18]))
print('living_area > 42:', len(data[data['living_area'] > 42]))
data_area = data[['total_area', 'living_area', 'kitchen_area', 'studio', 'is_apartment']]
print('Площадь кухни > жилой площади:', len(data_area[data_area['living_area'] < data['kitchen_area']]))
data_area[data_area['living_area'] < data['kitchen_area']].head()
```

```python
#циклом меняю значения living_area и kitchen_area местами, если kitchen_area > living_area
pd.set_option('chained_assignment', None)
for index in range(len(data)): 
    if data['living_area'][index] < data['kitchen_area'][index]:
        data['living_area'][index], data['kitchen_area'][index] = data['kitchen_area'][index], data['living_area'][index]
data_area = data[['total_area', 'living_area', 'kitchen_area', 'studio', 'is_apartment']]
print('Площадь кухни > жилой площади:', len(data_area[data_area['living_area'] < data['kitchen_area']]))
data[['total_area', 'living_area', 'kitchen_area', 'rooms', 'last_price']].corr()
```

```python
#категоризирую total_area делю на 30 равных групп
data['total_area'].describe()
data_area[data_area['total_area'] < 38].count()
data['total_area_id'] = pd.qcut(data['total_area'], q=30) 
data['total_area_id'].unique()
```

```python
#заполняю пропуски в living area и kitchen_area медианами группированных по категориям данных
data['living_area'] = data.groupby('total_area_id')['living_area'].apply(lambda s: s.fillna(s.median()))
print('Количество Nan в living_area', data['living_area'].isna().sum())
data['living_area'].hist(bins=15, range=(0,60), alpha=0.4)
data['kitchen_area'] = data.groupby('total_area_id')['kitchen_area'].apply(lambda s: s.fillna(s.median()))
print('Количество Nan в kitchen_area', data['kitchen_area'].isna().sum())
data['kitchen_area'].hist(bins=15, range=(0,60), alpha=0.4)
```

```python
data = data.dropna(subset=['locality_name']).reset_index(drop=True)

data['airports_nearest'].fillna(0, inplace=True) # заменяю на 0, чтобы было понятно что отсутствуют данные

#data['city_centers_nearest'].fillna(0, inplace=True) # заменяю на 0, чтобы было понятно что отсутствуют данные

print('Количество значений 0 в столбце parks_around_3000 =', len(data[data['parks_around_3000'] == 0]))
data['parks_around_3000'].fillna(0, inplace=True) # заменяю на 0, тк значение 0 встречается
print('Количество значений 0 в столбце parks_nearest =', len(data[data['parks_nearest'] == 0]))
data['parks_nearest'].fillna(0, inplace=True) # заполняю 0, как отсутствующие значения
print('Количество значений 0 в столбце ponds_around_3000 =', len(data[data['ponds_around_3000'] == 0]))
data['ponds_around_3000'].fillna(0, inplace=True) # заполняю 0, как отсутствующие значения
print('Количество значений 0 в столбце ponds_nearest =', len(data[data['ponds_nearest'] == 0]))
data['ponds_nearest'].fillna(0, inplace=True)
print('Количество значений 0 в столбце days_exposition =', len(data[data['days_exposition'] == 0]))
data['days_exposition'].fillna(0, inplace=True) # заполняю 0, как отсутствующие значения
```

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition']) #изменяю тип данных на datetime
data.head()
```

<br> parks_around3000 тип float, количество парков в радиусе 3000м не может быть дробным числом, поэтому переведу в int
<br> parks_nearest тип float, считаю что будет достаточно точности целых чисел, поэтому переведу в int

## Расчёты и добавление результатов в таблицу

```python
#Расчет стоимости квадратного метра уже добавлены ранее в столбец 'a_price'.
#Добавляю столбцы с днем недели, месяцем и годом публикации объявления
data['week_day'] = data['first_day_exposition'].dt.strftime('%A')
data['month'] = data['first_day_exposition'].dt.strftime('%B')
data['year'] = data['first_day_exposition'].dt.strftime('%Y')
data.head()
```

```python
# функция категоризации по этажам
def floor_id(row): 
    floor = row['floor']
    last = row['floors_total']
    if floor == 1:
        return 'first'
    elif floor == last:
        return 'last'
    else:
        return 'other'
# применяю функцию и категоризирую данные по этажу    
data['floor_id'] = data.apply(floor_id, axis=1) 
data.head()
```
